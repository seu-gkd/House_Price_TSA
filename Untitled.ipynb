```python
import pandas as pd
from fbprophet import Prophet
```

```python
def clean(file):
    with open('target.txt','a+',encoding='utf-8') as target:
        for line in open(file,'r',encoding='utf-8'):
            # 直辖市,北京,朝阳,2010年12月房价,27206元/㎡,1.41%↑
            data = line.split(',')
            # 地理位置
            # loc = data[0] + ',' + data[1] + ',' + data[2]
            loc = data[0] + ',' + data[1]
            # 日期
            # date = data[3].split('月房价')[0].replace('年','-')
            date = data[2].split('月房价')[0].replace('年','-')
            # 价格
            # price = data[4].split('元/㎡')[0]
            price = data[3].split('元/㎡')[0]
            target.write(loc + ',' + date + ',' + price + '\n')
```

```python
clean('test_info.txt')
```

```python
data = pd.read_csv("target.txt")
print(data)
```

```python
data['日期'] = pd.to_datetime(data['日期'])
data.columns = ['省','市','ds','y']
print(data)
```

```python
m = Prophet()
# m.add_country_holidays(country_name='CN')
m.fit(data[data['市'] == '常州'])
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```
